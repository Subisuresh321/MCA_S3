```python
x=np.array([1,5,3,3,3,3])
y=np.array([3,3,3,1,3,5])
plt.plot(x,y,marker='+',color="r",ms=5,mec='r',mfc='g')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a1 = np.array([1, 2, 3, 4])
a2 = np.array([5, 6, 7, 8])
x = np.array([0, 1, 2, 3])

plt.plot(x, a1, marker=".", label="a1")
plt.plot(x, a2, marker=".", label="a2")
plt.show()
```

```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([5, 6, 7, 8])

plt.subplot(2,2,1)
plt.plot(a1)
plt.xlabel("Time")
plt.ylabel("Distance")
plt.subplot(2,2,2)
plt.plot(a2)
plt.subplot(2,2,3)
plt.plot(a1)
plt.subplot(2,2,4)
plt.plot(a2)
plt.show()
```
